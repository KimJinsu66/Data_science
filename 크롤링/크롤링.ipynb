```python
import requests
from bs4 import BeautifulSoup # bs4 중에 BeautifulSoup만 불러오겠다는 의미이다.

req = requests.get('http://tv.naver.com/r/')
raw = req.text # 가저온것은 HTML 타입이 아니다 .String 타입의 일반적인 텍스트에 불과하다

html = BeautifulSoup(raw,'html.parser') # BeautifulSoup는 String 값을 살아있는 HTML문서로 바꿔줌

infos = html.select('div.cds') # select 함수 쓰면 선택자를 갖는 요소를 리스트타입으로 가져온다.
clip1 = infos[0]
clip1_title = clip1.select_one('dt.title tooltip') # tooltip 자료 추출
print(clip1_title) 
print(clip1_title.text) # 출력하면 차이점 알수 있다.
print(infos[0].select_one('dt.title tooltip').text) # 위에 코드를 한줄로 정리 
```
