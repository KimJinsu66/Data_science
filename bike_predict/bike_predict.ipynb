```python
#원 데이터는 방금 그래프와 같지만 모형간의 비교를 위해 데이터를 조작한다. 
#원데이터에서 30 이상의 값들을 모두 30으로 설정했다.
df["cliped_rate"] = df.rate.clip_upper(30)
df.cliped_rate.plot()
plt.title("호흡기 질환 사망자 수")
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm

model_ARMA = sm.tsa.ARMA(df["cliped_rate"], (1,2))
result_ARMA = model_ARMA.fit()
print(result_ARMA.summary())
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(df, order=(0,1,1))
model_fit = model.fit(trend='c',full_output=True, disp=1)
print(model_fit.summary())
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(df, order=(0,1,1))
model_fit = model.fit(trend='nc',full_output=True, disp=1)
print(model_fit.summary())
```
