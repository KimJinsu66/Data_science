```python
%matplotlib inline 
# Shift + Enter을 차지 않고 Run버튼을 눌러 실행하는 그래프를 얻을 수 있다
from matplotlib import pyplot as plt

plt.plot([1,2,3],[110,130,120])
plt.show()
```

```python
# 제목과 축 레이블 
 
plt.plot(["Seoul","Paris","Seattle"], [30,25,55])
plt.xlabel('City')
plt.ylabel('Response') # 답장
plt.title('Experiment Result') # 실험결과
plt.show()
```

```python
# 플롯에 여러개 라인들 추가하기 위해서는 plt.plot을 여러개 사용
# legend 함수 사용

plt.plot([1,2,3], [1,4,9])
plt.plot([2,3,4],[5,6,7])
plt.xlabel('Sequence')
plt.ylabel('Time(secs)')
plt.title('Experiment Result')
plt.legend(['Mouse', 'Cat'])
plt.show()
```
