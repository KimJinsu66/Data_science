```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10)
y = x*5

plt.plot(x,y,'g')# 3번째 파라미터 색깔을 바꿔준 ex)b= blue, r = red
plt.show()
```

```python
# 선종류 변경하기

x = np.arange(1,10)
y = x*5

plt.plot(x,y,'or')# or 선 종류 변경하기 동그란 마커를 o로 표시
plt.show()
```

```python
# 라벨과 타이틀
x = np.arange(1,10)
y = x*5

plt.plot(x,y,'og')# 3번째 파라미터 색깔을 바꿔준 ex)b= blue, r = red
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.show()
```

```python
# 구간 확대/축소

x = np.arange(1,10)
y = x*5

plt.xlim(2,3) # x 2~3만 표시
plt.ylim(5,20) # y 5~20까지만 표시

plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.show()
```

```python
#레전드
"""
그래프를 그릴때 여러개의 그래프 같이 그릴 수 있다.
"""

x = np.arange(1,10,0.1)
y = x*0.2
y2 = np.sin(x) # ??



plt.plot(x,y,'b',label = "first")
plt.plot(x,y2,'r',label = "second")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.legend(loc = 'upper right')
plt.show()
```
