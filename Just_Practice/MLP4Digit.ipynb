```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

digits = datasets.load_digits()

x = digits.images.reshape((digits.images.shape[0], digits.images.shape[1]*digits.images.shape[2]))
y = digits.target

x_train = x[:1000]
print(x_train)
y_train = y[:1000]
print(y_train)
x_test = x[1000:]
y_test = y[1000:]
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(15,), 
                    activation='logistic', 
                    solver='sgd', 
                    tol=1e-4, 
                    learning_rate_init=.1, 
                    verbose=False)

mlp.fit(x_train,y_train)
```

```python
predictions = mlp.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, predictions)
```

```python
import pandas as pd

pd.crosstab(y_test, predictions, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
print(mlp.coefs_[0])# coefs = 계수
```

```python
print(mlp.intercepts_[0]) # 0번째 intercepts(절편)
print(mlp.intercepts_[1]) # 1번째 intercepts(절편)
```

```python
import pandas as pd
s_data = pd.read_csv('simpsons.csv') # csv에서 읽는다.
s_data['gender'] = s_data['gender'].map({'M':0,'F':1}) # gender에서 M은 0 F 1로 바꿔준다.
#print(s_data)

s_data = s_data.drop('Unnamed: 0', axis = 1) #이름을 떨궈낸다. axix 1 ?
x = s_data.drop('gender', axis = 1) # x는 drop을 떨궈낸 나머지 데이터
y = s_data['gender'] # y는 gender(결과값) 데이

print(x)
print(y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3) # 최대 깊이 3이다.

dtc.fit(x,y) #x,y를받아 fit 함수 분류기를 만들어 준다.
```

```python
from sklearn.tree import export_graphviz
export_graphviz(dtc, out_file = 'decision_tree.dot', #file 의 이름 
                feature_names = x.columns, # 특질 이름 ?? 
               class_names = ['M',"F"], 
               rounded = True)
# decision_tree.dot 파일이 생성된다.
```

```python
from sklearn import datasets

iris = datasets.load_iris()
print(dir(iris))
print(iris.feature_names)
print(iris.target_names)
```

```python
import pandas as pd

data =pd.DataFrame({
    'sepal length':iris.data[:,0],
    'sepal width':iris.data[:,1],
    'petal length':iris.data[:,2],
    'petal width':iris.data[:,3],
    'species':iris.target
})
data.head()
```

```python
from sklearn.model_selection import train_test_split

X = data[['sepal length','sepal width', 'petal length','petal length','petal width']]
y = data['species']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = clf.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
from sklearn import svm
#svm 사용 해봄
clf = svm.SVC(gamma = 'auto', kernel = 'rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
confusion_matrix(y_test,y_pred)
```
